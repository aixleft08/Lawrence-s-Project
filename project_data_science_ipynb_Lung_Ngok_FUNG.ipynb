# **High Level Overview: 2020**
Firstly, I will do an overview of the entire year. I created a new dataframe with my date_streamed grouped by
date and number of songs listened on each date.

```python
# create new df with number of songs listened per date
date_count_df = df_2020.groupby('date_streamed')[['date_streamed']].count()
#rename col to count
date_count_df.columns = ['Count']
# reset_index I can use 'date_streamed' to index into df
date_count_df.reset_index(inplace = True)
# view snapshot of df
date_count_df.head(5)
```

I created a scatter plot to help visualize the distribution of songs listened throughout 2020. 

```python
# scatter plot
# adjusts the size of the plot
plt.figure(figsize=(11,7))
# defines the type of plot and the data)
plt.scatter(
date_count_df["date_streamed"],
date_count_df["Count"]
)
# Adds a label to the x-axis
plt.xlabel('Time')
# Rotate the x-axis labels
plt.xticks(rotation=45)
# Adds a title to the y-axis
plt.ylabel('# of Songs Listened')
# Adds a title to the entire plot
plt.title('Songs Listened Throughout 2020')
# Displays Plot
plt.show()
```

Unsurprisingly, my songs listened drastically did not appear over the summer. This is because of my travel back to Shenzhen. I was using another music app when I was in Shenzhen.
With this in mind, I then investigated by my breakdown of number of songs listened per category for the entire
year. I sorted it in decending order by number of songs listened.

```python
genre_count_df = df_2020.groupby('genre')[['date_streamed']].count().sort_values(by='date_streamed', ascending=False)
genre_count_df

genre_count_df.rename(columns={'date_streamed':'number of songs listened'})
```

I focused only on the top 7 genres, as the others have less than 40 total songs listened. For the genres
that have less than 40 total songs listened, I created a new section called 'other'. I then create a pie chart to
easily visualize my songs listened breakdown by percentage.

```python
genre_count_df_sorted = genre_count_df[0:7]
genre_count_df_sorted.reset_index(inplace=True)
genre_count_df_sorted

# sum number of songs listned for genres with less than 5 songs listened
other_total = genre_count_df['date_streamed'][7:].sum()

# add other genre to sorted df at the last row
genre_count_df_sorted.loc[len(genre_count_df_sorted)] = ['other', other_total]

genre_count_df_sorted.rename(columns={'date_streamed':'number of songs listened'},inplace=True)
genre_count_df_sorted
```

```python
# pie chart
colors = sns.color_palette('pastel')[0:8]
plt.figure(figsize=(10, 6), dpi=80)

plt.pie(
genre_count_df_sorted['number of songs listened'],
labels = genre_count_df_sorted['genre'],
colors = colors,
autopct='%1.1f%%')
plt.title('# of songs listened per genre for 2020')
plt.show()
```

## K-POP

As k-pop takes up the biggest percetnage, I decied to do an in-depth breakdown and see if I could discover
any new patterns.

Firstly I created two separate dataframes for just k-pop.

```python
kpop_df = df_2020.query("genre == 'k-pop'")
kpop_df.head()
```

Next, I looked into how my habits changed over the course of the year, by month. I decided to organize the
data by month to hopefully better recogonize patterns than if it was by day.

```python
kpop_month_df = kpop_df.groupby(['month_name','month'])[['date_streamed']].count()
kpop_month_df_sorted = kpop_month_df.sort_values(by='month')

# reset_index so i can use column names
# inplace so i don't have to create a new df
kpop_month_df_sorted.reset_index(inplace=True)
kpop_month_df_sorted
```

```python
# bar chart

# plot size
plt.figure(figsize=(10,6))

# plot type and data
plt.bar(
kpop_month_df_sorted['month_name'],
kpop_month_df_sorted['date_streamed'])

# labels
plt.xticks(rotation=90)
plt.title('number of kpop songs listened by month')
plt.ylabel('# of songs listened')
plt.xlabel('month')

# display plot
plt.show()
```
